```python
#Loading the required libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Loading the IPL matches dataset
ipl=pd.read_csv("matches.csv")
```

```python
#Haaving a glance at first five records of the dataset
ipl.head()
```

```python
#Looking at the number of rows and columns in the dataset
ipl.shape
```

```python
#Getting the frequency of most man of the match awards
ipl["player_of_match"].value_counts()
```

```python
#Getting the top 10 players with most man of the match awarrds
ipl["player_of_match"].value_counts()[0:10]
```

```python
#Getting the top 5 players with most man of the match awards
ipl["player_of_match"].value_counts()[0:5]
```

```python
#Make a bar plot corresponding to the above code cell
plt.bar(list(ipl["player_of_match"].value_counts()[0:5].keys()),list(ipl["player_of_match"].value_counts()[0:5]))
plt.title("Bar Plot For Top 5 Players With Most Man Of The Match Awards")
plt.xlabel("No. Of Man Of Match Awards")
plt.ylabel("Player Names")
plt.show()
```

```python
sns.barplot(x=list(ipl["player_of_match"].value_counts()[0:5].keys()),y=list(ipl["player_of_match"].value_counts()[0:5]))
plt.show()
```

```python
#Getting the frequency of result column
ipl["result"].value_counts()
```

```python
#Finding out the number of toss wins w.r.t each team
ipl["toss_winner"].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first=ipl[ipl["win_by_runs"]!=0]
```

```python
#Looking at the head of above dataset
batting_first.head()
```

```python
#Making a histogram of the above dataset
plt.hist(batting_first["win_by_runs"])
plt.title("Distribution Of Margins By Which Team Batting First Won")
plt.xlabel("No. Of Runs")
plt.ylabel("Count")
plt.show()
```

```python
sns.displot(batting_first["win_by_runs"])
plt.show()
```

```python
#Finding out the number of wins  w.r.t each team after batting first
batting_first["winner"].value_counts()
```

```python
#Making a bar plot for top 3 teams with most wins after batting first
plt.bar(batting_first["winner"].value_counts()[0:3].keys(),batting_first["winner"].value_counts()[0:3])
plt.title("Bar Plot For Most Wins After Batting First")
plt.xlabel("Team Name")
plt.ylabel("Count")
plt.show()
```

```python
sns.barplot(x=batting_first["winner"].value_counts()[0:3].keys(),y=batting_first["winner"].value_counts()[0:3])
plt.show()
```

```python
#Making a pie chart for most number of wins w.r.t each team after batting first
plt.pie(batting_first["winner"].value_counts(),labels=batting_first["winner"].value_counts().keys(),autopct="%0.1f%%")
plt.show()
```

```python
#Extracting those records where a team won after batting second
batting_second=ipl[ipl["win_by_wickets"]!=0]
```

```python
#Looking at the head of above dataset
batting_second.head()
```

```python
sns.displot(batting_second["win_by_wickets"])
plt.show()
```

```python
#Finding out the frequency of number of wins w.r.t each team after batting second
batting_second["winner"].value_counts()
```

```python
#Making a bar plot for top 3 teams with most wins after batting second
plt.bar(batting_second["winner"].value_counts()[0:3].keys(),batting_second["winner"].value_counts()[0:3])
plt.title("Bar Plot For Most Wins After Batting Second")
plt.xlabel("Team Name")
plt.ylabel("Count")
plt.show()
```

```python
sns.barplot(x=batting_second["winner"].value_counts()[0:3].keys(),y=batting_second["winner"].value_counts()[0:3])
plt.show()
```

```python
#Making a pie chart for most number of wins w.r.t each team after batting second
plt.pie(batting_second["winner"].value_counts(),labels=batting_second["winner"].value_counts().keys(),autopct="%0.1f%%")
plt.show()
```
